```python
import miceforest as mf # cargamos la libreria miceforest
from sklearn.datasets import load_iris # Extraigo el datasets iris
import numpy as np #  importamos numpy, es para calculos numericos
```

```python
#from sklearn import datasets
iris = load_iris()
print(iris)
```

```python
pd.DataFrame(iris.data)
```

```python
type(iris)
```

```python
# Convertimos del array sklearn a dataframe

iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target  # Add the target variable as a column
```

```python
iris_df.columns
```

```python
# renombramos la columna target en species
iris_df.rename({'target':'species'},inplace=True,axis=1)
```

```python
iris_df
```

```python
# transformo el tipo de dato categorica
iris_df['species'] = iris_df['species'].astype('category')
iris_df.info()
```

```python
# asigno valores nulos a las columnas aleatoriamente
iris_amp = mf.ampute_data(iris_df,perc=0.5,random_state=1991)# realizo una amputacion de datos asignado NaN
iris_amp
# random_state= 1991 profundidad de arbol
# perc= 0.5 25% de datos nulos
```

```python
# Crear Kernel
kds = mf.ImputationKernel(
  iris_amp,          #Nombre de dataframe para entrenar modelo
  random_state=1991  # valores aleatorios para entrenar modelo
)
# correr el algoritmo MICE durante 2 iteraciones
kds.mice(2)
#Retorna el dataset completo

iris_completo = kds.complete_data()
iris_completo
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el archivo Excel
file_path = 'E:\Analisis de Datos\Data\Irena\IRENA_Stats_extract_2024 H2.xlsx'
xls = pd.ExcelFile(file_path)
#E:\Analisis de Datos\Data\Irena

# Cargar la hoja 'Region'
region_data = pd.read_excel(xls, sheet_name='Region')

# Filtrar solo energías renovables
renewables = region_data[region_data['RE or Non-RE Group'] == 'Total Renewable']

# Agrupar por región y sumar la capacidad instalada más reciente
latest_year = renewables['Year'].max()
renewable_by_region = (renewables[renewables['Year'] == latest_year]
                       .groupby('Region')['Sum of Electricity Installed Capacity (MW)']
                       .sum()
                       .sort_values(ascending=False))

# Visualizar las regiones líderes
plt.figure(figsize=(10, 6))
renewable_by_region.plot(kind='bar', color='skyblue')
plt.title(f'Capacidad Instalada de Energías Renovables por Región ({latest_year})')
plt.xlabel('Región')
plt.ylabel('Capacidad Instalada (MW)')
plt.xticks(rotation=45)
plt.show()
```
